```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
import torch
```

```python
x = np.linspace(0,10, 101)
```

```python
x
```

```python
def fx(value):
    return (3*(value**2))-(2*value)+5
```

```python
y = np.array([fx(i) for i in x])
```

```python
y
```

```python
df = pd.DataFrame({
    'x': x,
    'y': y
})
```

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel('y')
```

```python
tensor_x = torch.tensor(5.0, requires_grad=True)
```

```python
tensor_y = (3*(tensor_x**2))-(2*tensor_x)+5
tensor_y.backward()
```

```python
tensor_x.grad
```
