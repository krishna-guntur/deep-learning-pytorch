```python
from used_car_analysis import Used_Car_Analysis
import time

import torch, pandas as pd, numpy as np
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
```

```python
gpu = torch.device('cuda')

dataset = Used_Car_Analysis()
```

```python
X_train_np, X_test_np, y_train_ser, y_test_ser = dataset.get_dataset()

X_train = torch.tensor(np.array(X_train_np), dtype=torch.float32)
X_test  = torch.tensor(np.array(X_test_np),  dtype=torch.float32)

y_train = torch.tensor(np.array(y_train_ser), dtype=torch.float32).view(-1, 1)
y_test  = torch.tensor(np.array(y_test_ser),  dtype=torch.float32).view(-1, 1)

```

```python
class Custom_Data(Dataset):

    def __init__(self, features, labels):
        super().__init__()

        self.features = features
        self.labels = labels

    def __len__(self):

        return self.features.shape[0]
    
    def __getitem__(self, index):

        return self.features[index], self.labels[index]
```

```python
train_data = Custom_Data(X_train, y_train)
test_data = Custom_Data(X_test, y_test)

train_loader = DataLoader(train_data, batch_size=1000, shuffle=True)
test_loader = DataLoader(test_data, batch_size=1000, shuffle=True)
```

```python
class neural_net(nn.Module):

    def __init__(self, num_features):
        super().__init__()
        self.num_features = num_features
        self.linear1 = nn.Linear(self.num_features, 6)
        #self.relu = nn.ReLU()
        self.linear2 = nn.Linear(6, 3)
        #self.sigmoid1 = nn.Sigmoid()
        self.linear3 = nn.Linear(3,1)
        #self.sigmoid2 = nn.Sigmoid()

    def forward(self, features):
        out = self.linear1(features)
        #out = self.relu(out)
        out = self.linear2(out)
        #out = self.sigmoid1(out)
        out = self.linear3(out)
        #out = self.sigmoid2(out)

        return out
```

```python
learning_rate = 0.01
epoch_num = 25
```

```python
#creating a network
network = neural_net(X_train.shape[1])

#optimizer
optimizer = torch.optim.SGD(network.parameters(), learning_rate)

#loss function
loss_function = nn.MSELoss()
```

```python
for epoch in range(epoch_num):

    for batch_features, batch_labels in test_loader:

        start_time = time.time()
        y_hat = network(batch_features)
        loss = loss_function(y_hat, batch_labels.view(-1, 1))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        time_taken = start_time - time.time()

    print(f"Epoch {epoch+1}, Loss: {loss.item()}")
    print(time_taken)
```
